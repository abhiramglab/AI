```python
# Bredth first search in search of target - Using Brute Force Algorithms

def bfs(src,target):
    q = [src]
    visited_states = []
    while len(q):
        state = q.pop(0)
        print(state)
        visited_states.append(state)
        if state == target:
            print("Success")
            return
        for move in possible_moves(state, visited_states):
            if move not in q and move not in visited_states:
                q.append(move)
    print("Fail")
```

```python

def possible_moves(state,visited_states): 
    b = state.index(-1) 
    d = []
    if b + 3 in range(9): d.append('d')
    if b - 3 in range(9): d.append('u')
    if b not in [0,3,6]: d.append('l')
    if b not in [2,5,8]: d.append('r')
    pos_moves = []
    for m in d:
        pos_moves.append(gen(state, m, b))
    return [move for move in pos_moves if move not in visited_states]
```

```python
def gen(state, m, b): 
    temp = state.copy()
    if m == 'd': temp[b+3] , temp[b] = temp[b], temp[b+3]
    if m == 'u': temp[b-3] , temp[b] = temp[b], temp[b-3]
    if m == 'l': temp[b-1] , temp[b] = temp[b], temp[b-1]
    if m == 'r': temp[b+1] , temp[b] = temp[b], temp[b+1]
    return temp
```
